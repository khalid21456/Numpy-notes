# Random Permutations

**The <font color="red">shuffle()</font> method makes changes to the original array.**


```python
arr = np.array([1, 2, 3, 4, 5])
random.shuffle(arr)
print(arr)
```

**The <font color="red">permutation()</font> method returns a re-arranged array (and leaves the original array un-changed).**

```python
arr = np.array([1, 2, 3, 4, 5])
print(random.permutation(arr))
print(arr)
```

# Normal (Gaussian) Distribution

```python
x = random.normal(loc = 0, scale = 1 , size=(10))
print(x)
```

```python
sns.distplot(random.normal(size=50), hist=False)
plt.show()
```

**<font color="red">Note: </font>The curve of a Normal Distribution is also known as the Bell Curve because of the bell-shaped curve.**

# Binomial Distribution

```python
x = random.binomial(n=10, p=0.5, size=10)
print(x)
```

```python
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)
plt.show()
```

# Difference Between Normal and Binomial Distribution

**The main difference is that normal distribution is continous whereas binomial is discrete, but if there are enough data points it will be quite similar to normal distribution with certain loc and scale.**

```python
sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')

plt.show()
```
